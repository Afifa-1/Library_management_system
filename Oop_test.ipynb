```python
# Encapsulation

class book:
    def __init__ (self,title,author,isbn):
        self.__title = title
        self.__author = author
        self.__isbn = isbn

    def get_title(self):
        return self.__title
        
    def set_title(self,title):
        self.__title  = title
        return self.__title
        
    def get_author(self):
        return self.__author
        
    def set_title(self,author):
        self.__author  = author
        return self.__author
    
    def get_isbn(self):
        return self.__isbn
        
    def set_isbn(self,isbn):
        self.__isbn = isbn
        return self.__isbn
        
    def display_info(self):
        print(f"Book Details: {self.__title},{self.__author}, {self.__isbn}")


# Inheritance

class Ebook(book):
    def __init__(self,title,author,isbn, file_size):
        super().__init__(title,author,isbn)
        self.file_size = file_size

class printed_books(book):
    def __init__ (self,title,author,isbn,pages):
        super().__init__(title,author,isbn)
        self.pages = pages



my_book = Ebook("python" , "Jay jay", 987-678 , "10MBS")

my_book.display_info()


# Abstraction

from abc import ABC,abstractmethod

class person(ABC):
    @abstractmethod
    def role (self):
        pass

class Member(person):
    def role(self):
        return "Member"
    def borrow_book(self, book):
        print(f"person is borrowing : {book.get_title()}")

    def return_book(self):
        print("person is returning the book")

class Librarian(person):

    def role(self):
        return "Librarian"

    def issue_book(self,book,member_name):
        print(f"librarian is issuing book : {book.get_title()} to {member_name}")

    def check_availability(self,book):
        print(f"{book.get_title()} Book is available in the library")


# Polymorphism

def get_person_role(person):
    print(f"Role:{person.role()}")

def borrow_book_from_library(book,person):
    if person.role() == "Member":
        person.borrow_book(book)
    elif person.role() == "Librarian":
        person.issue_book (book , "john doe") 



staff = Librarian()
user = Member()

(get_person_role(user))
(get_person_role(staff))



(staff.check_availability(my_book))

borrow_book_from_library (my_book, user)
borrow_book_from_library (my_book, staff)
```
